```python
#clculation with function
def calculator():
    def add(n1, n2):
        return n1 + n2

    def substract(n1, n2):
        return n1 - n2

    def multiply(n1, n2):
        return n1 * n2

    def devide(n1, n2):
        return n1 / n2

    operations = {
        "+": add,
        "-": substract,
        "*": multiply,
        "/": devide
    }

    num_1 = float(input("Put your first number: "))
    num_2 = float(input("Put your second number: "))
    loop = True

    while loop:
        for symbol in operations:
            print(symbol)
        pick_symbol = input("Pick an operation from the line above: ")
        
        if pick_symbol not in operations:
            print("Invalid operation.")
            continue

        answer1 = operations[pick_symbol](num_1, num_2)
        print(f"{num_1} {pick_symbol} {num_2} = {answer1}")
        
        if input("Write 'ye' or 'no' to show if you want to continue the calculations: ") == "ye":
            num_1 = answer1
            num_2 = int(input("Put your next number: "))
        else:
            loop = False

# Call the calculator function
calculator()
```
